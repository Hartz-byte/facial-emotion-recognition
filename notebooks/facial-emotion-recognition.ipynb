## Imports

```python
# Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, GlobalAveragePooling2D, Conv2D, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
```

## Load & Visualize Dataset

```python
# Load & Visualize Dataset
data_dir = "../data"
train_dir = os.path.join(data_dir, "train")
test_dir = os.path.join(data_dir, "test")

emotion_labels = sorted(os.listdir(train_dir))

# Count class distribution
def count_images(directory):
    counts = {emotion: len(os.listdir(os.path.join(directory, emotion))) for emotion in emotion_labels}
    return pd.Series(counts)

train_counts = count_images(train_dir)

# Plot class distribution
plt.figure(figsize=(10, 4))
sns.barplot(x=train_counts.index, y=train_counts.values)
plt.title("Training Set - Emotion Distribution")
plt.ylabel("Number of Images")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

## Image Preprocessing and Augmentation

```python
# Image Preprocessing and Augmentation
def preprocess_input_efficientnet(x):
    return (x / 255.0 - 0.5) * 2.0

IMG_SIZE = 96
BATCH_SIZE = 32

data_gen_args = dict(
    preprocessing_function=preprocess_input_efficientnet,
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    brightness_range=(0.8, 1.2),
    validation_split=0.2
)

train_aug = ImageDataGenerator(**data_gen_args)
test_aug = ImageDataGenerator(preprocessing_function=preprocess_input_efficientnet)

train_data = train_aug.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training",
    shuffle=True
)

val_data = train_aug.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
    shuffle=True
)

test_data = test_aug.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)
```

## EfficientNetB0-based Model

```python
# Model definition
def build_model():
    input_layer = Input(shape=(IMG_SIZE, IMG_SIZE, 1))
    x = Conv2D(3, (3,3), padding='same')(input_layer)  # grayscale to RGB
    base_model = EfficientNetB0(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base_model(x, training=False)
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.4)(x)
    output_layer = Dense(7, activation='softmax', kernel_regularizer=l2(0.01))(x)
    model = Model(inputs=input_layer, outputs=output_layer)
    return model, base_model

loss = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)
model = build_model()
model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])
model.summary()
```

## Callbacks

```python
# Create a directory if not exist
os.makedirs("saved_model", exist_ok=True)

callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint("saved_model/best_model.keras", monitor='val_accuracy', save_best_only=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-7, verbose=1)
]
```

## Class Imbalance

```python
# Class weights
labels = train_data.classes
class_weights = class_weight.compute_class_weight('balanced', classes=np.unique(labels), y=labels)
class_weights_dict = dict(enumerate(class_weights))
```

## Initial Training

```python
# Initial Training
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=30,
    callbacks=callbacks,
    class_weight=class_weights_dict
)
```

## Unfreezing and Fine-tuning EfficientNetB0

```python
# Unfreeze last few layers of EfficientNetB0
for layer in base_model.layers[-50:]:
    layer.trainable = True

# Compile again
model.compile(optimizer=tf.keras.optimizers.Adam(1e-5), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Model Fine Tuning Training
print("\n--- Starting Fine-Tuning ---\n")

history_finetune = model.fit(
    train_data,
    validation_data=val_data,
    epochs=60,
    initial_epoch=len(history.history['accuracy']),
    callbacks=callbacks,
    class_weight=class_weights_dict
)
```

## Evaluation

```python
# Evaluate on Test Data
loss, acc = model.evaluate(test_data)
print(f"Test Accuracy: {acc:.4f}")

# Predict
predictions = model.predict(test_data)
y_pred = np.argmax(predictions, axis=1)
y_true = test_data.classes
```

## Confusion matrix

```python
# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=emotion_labels, yticklabels=emotion_labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## Classification report

```python
# Classification report
print(classification_report(y_true, y_pred, target_names=emotion_labels))
```

## Training History

```python
# Plot Training History
plt.figure(figsize=(8,4))
plt.plot(history.history['accuracy'], label="Train Accuracy")
plt.plot(history.history['val_accuracy'], label="Val Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.title("Initial Training Accuracy")
plt.show()

# Plot Fine Tuning Training History
plt.figure(figsize=(8,4))
plt.plot(history_finetune.history['accuracy'], label="Train Accuracy")
plt.plot(history_finetune.history['val_accuracy'], label="Val Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.title("Fine-Tuning Accuracy")
plt.show()
```

## t-SNE visualization

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder

# Step 1: Extract Features Before Final Dense Layer
feature_extractor = tf.keras.Model(
    inputs=model.input,
    outputs=model.layers[-3].output
)

features = feature_extractor.predict(test_data, verbose=1)
labels = test_data.classes
preds = np.argmax(model.predict(test_data), axis=1)

# Step 2: Reduce Features to 2D using t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42, n_iter=1000)
features_2d = tsne.fit_transform(features)

# Step 3: Visualize
plt.figure(figsize=(10, 6))
scatter = plt.scatter(features_2d[:, 0], features_2d[:, 1], c=labels, cmap="tab10", alpha=0.6)
plt.legend(handles=scatter.legend_elements()[0], labels=emotion_labels, title="Emotions", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("t-SNE Visualization of Test Features")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.grid(True)
plt.tight_layout()
plt.show()

# Highlight Incorrect Predictions
incorrect = preds != labels
plt.figure(figsize=(10, 6))
plt.scatter(features_2d[~incorrect, 0], features_2d[~incorrect, 1], c='green', alpha=0.4, label='Correct')
plt.scatter(features_2d[incorrect, 0], features_2d[incorrect, 1], c='red', alpha=0.6, label='Incorrect')
plt.legend()
plt.title("t-SNE: Correct vs Incorrect Predictions")
plt.grid(True)
plt.tight_layout()
plt.show()
```
