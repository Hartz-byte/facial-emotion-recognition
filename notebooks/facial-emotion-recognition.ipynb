## Imports

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, GlobalAveragePooling2D, Conv2D
from tensorflow.keras.regularizers import l2
import cv2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

## Load & Visualize Dataset

```python
# Load & Visualize Dataset
data_dir = "../data"
train_dir = os.path.join(data_dir, "train")
test_dir = os.path.join(data_dir, "test")

emotion_labels = sorted(os.listdir(train_dir))

def count_images(directory):
    counts = {emotion: len(os.listdir(os.path.join(directory, emotion))) 
              for emotion in emotion_labels}
    return pd.Series(counts)

train_counts = count_images(train_dir)
test_counts = count_images(test_dir)

# Plot class distribution
plt.figure(figsize=(10,4))
sns.barplot(x=train_counts.index, y=train_counts.values)
plt.title("Training Set - Emotion Distribution")
plt.ylabel("Number of Images")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

## Image Preprocessing and Augmentation

```python
# Image Preprocessing and Augmentation
# IMG_SIZE = 48
IMG_SIZE = 96
BATCH_SIZE = 32

def preprocess_input_efficientnet(x):
    return (x / 255.0 - 0.5) * 2.0

train_aug = ImageDataGenerator(
    preprocessing_function=preprocess_input_efficientnet,
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    brightness_range=(0.8, 1.2),
    validation_split=0.2
)

test_aug = ImageDataGenerator(
    preprocessing_function=preprocess_input_efficientnet
)

train_data = train_aug.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training",
    shuffle=True
)

val_data = train_aug.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
    shuffle=True
)

test_data = test_aug.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)
```

## EfficientNetB0-based Model

```python
# EfficientNetB0-based Model
def build_model():
    input_layer = Input(shape=(IMG_SIZE, IMG_SIZE, 1))
    x = Conv2D(3, (3,3), padding='same')(input_layer)
    # base_model = ResNet50(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base_model = EfficientNetB0(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base_model(x, training=False)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.4)(x)
    output_layer = Dense(7, activation='softmax', kernel_regularizer=l2(0.01))(x)
    return Model(inputs=input_layer, outputs=output_layer)

loss = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1)

model = build_model()
model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])
model.summary()
```

## Callbacks

```python
# Create a directory if not exist
os.makedirs("saved_model", exist_ok=True)

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint("saved_model/best_model.keras", monitor='val_accuracy', save_best_only=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-7, verbose=1)
]
```

## Class Imbalance

```python
# Class Imbalance
labels = train_data.classes
class_weights = class_weight.compute_class_weight('balanced', classes=np.unique(labels), y=labels)
class_weights_dict = dict(enumerate(class_weights))
```

## Model Training

```python
# Model Training
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=30,
    callbacks=callbacks,
    class_weight=class_weights_dict
)
```

## Unfreezing and Fine-tuning EfficientNetB0

```python
# Unfreeze last few layers of EfficientNetB0
for layer in base_model.layers[-50:]:
    layer.trainable = True

# Compile again (lower LR)
model.compile(optimizer=tf.keras.optimizers.Adam(1e-6), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Model Fine Tuning Training
print("\n--- Starting Fine-Tuning ---\n")

history_finetune = model.fit(
    train_data,
    validation_data=val_data,
    epochs=30,
    callbacks=callbacks,
    class_weight=class_weights_dict
)
```

## Evaluation

```python
# Evaluate on Test Data
loss, acc = model.evaluate(test_data)
print(f"Test Accuracy: {acc:.4f}")
```

```python
# Predict
predictions = model.predict(test_data)
y_pred = np.argmax(predictions, axis=1)
y_true = test_data.classes

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=emotion_labels, yticklabels=emotion_labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Classification Report
print(classification_report(y_true, y_pred, target_names=emotion_labels))
```

## Training History

```python
# Plot Training History
plt.figure(figsize=(8,4))
plt.plot(history.history['accuracy'], label="Train Accuracy")
plt.plot(history.history['val_accuracy'], label="Val Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.title("Accuracy Over Epochs")
plt.show()

# Plot Fine Tuning Training History
plt.figure(figsize=(8,4))
plt.plot(history_finetune.history['accuracy'], label="Train Accuracy")
plt.plot(history_finetune.history['val_accuracy'], label="Val Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.title("Accuracy Over Epochs")
plt.show()
```

## Grad-CAM Visualization

```python
import random

# Grad-CAM for internal ResNet layer
def get_gradcam(full_model, img_array_gray, class_idx, layer_name='conv5_block3_out'):
    # Convert grayscale to RGB for ResNet
    img_rgb = tf.repeat(img_array_gray, 3, axis=-1)

    # Get ResNet base model
    resnet_model = full_model.get_layer('resnet50')
    conv_output = resnet_model.get_layer(layer_name).output

    # Build the tail (everything after ResNet)
    x = conv_output
    # Go through layers after ResNet
    resnet_output_tensor = resnet_model.output
    found = False
    for layer in full_model.layers:
        if found:
            x = layer(x)
        if layer.output == resnet_output_tensor:
            found = True

    # Create a new model: from ResNet input → conv layer output → final prediction
    grad_model = tf.keras.models.Model(inputs=resnet_model.input, outputs=[conv_output, x])

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_rgb)
        loss = predictions[:, class_idx]

    grads = tape.gradient(loss, conv_outputs)
    if grads is None:
        raise ValueError("Gradients returned None — check model connection.")

    # Compute Grad-CAM
    grads = grads[0]
    conv_outputs = conv_outputs[0]
    weights = tf.reduce_mean(grads, axis=(0, 1))
    cam = tf.reduce_sum(weights * conv_outputs, axis=-1)

    # Normalize
    cam = tf.maximum(cam, 0)
    heatmap = cam / tf.reduce_max(cam)
    heatmap = cv2.resize(heatmap.numpy(), (img_array_gray.shape[2], img_array_gray.shape[1]))

    return heatmap


# Select random test image
batch_index = random.randint(0, len(test_data) - 1)
img_batch, label_batch = test_data[batch_index]

img_index = random.randint(0, img_batch.shape[0] - 1)
img = img_batch[img_index]
label = label_batch[img_index]
img_input = img.reshape(1, IMG_SIZE, IMG_SIZE, 1)

# Predict using the full model
pred = model.predict(img_input)
pred_class = np.argmax(pred)

# From RGB input directly through ResNet (just for sanity check)
resnet_model = model.get_layer('resnet50')
img_rgb = tf.repeat(img_input, 3, axis=-1)
resnet_pred = resnet_model(img_rgb)
print("ResNet-only prediction shape:", resnet_pred.shape)
print("Internal predictions shape:", pred.shape)
print("Requested class index:", pred_class)

# Grad-CAM using ResNet submodel and RGB version of the input
heatmap = get_gradcam(model, img_input, pred_class)

# Visualization
plt.figure(figsize=(6, 3))

plt.subplot(1, 2, 1)
plt.imshow(img_input[0].squeeze(), cmap='gray')
plt.title("Original")

plt.subplot(1, 2, 2)
plt.imshow(img_input[0].squeeze(), cmap='gray')
plt.imshow(heatmap, cmap='jet', alpha=0.5)
plt.title("Grad-CAM")
plt.tight_layout()
plt.show()
```
